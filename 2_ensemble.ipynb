```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans, AffinityPropagation, MeanShift, SpectralBiclustering, SpectralClustering, AgglomerativeClustering, DBSCAN, OPTICS, Birch

from sklearn.mixture import GaussianMixture
```

```python
# Importing the data

data = pd.read_csv("data/data.csv")
submission = pd.read_csv("data/sample_submission.csv")
```

```python
# Dropping ID column
data.drop(columns = 'id', inplace=True)
```

```python
# MiniBatch KMeans, Affinity Propagation, MeanShift, Spectral Clustering, Ward, Agglomerative Clustering, DBSCAN, OPTICS, BIRCH, Gaussian Mixture
sample = data

clusters_k_means = KMeans(n_clusters = 7).fit_predict(sample)
clusters_mixture = GaussianMixture(n_components = 7).fit_predict(sample)
```

```python
from scipy.sparse import csr_matrix

def create_sparse_matrix(clusters):
    n = len(clusters)
    data = []
    row = []
    col = []
    for i in range(n):
        for j in range(i+1, n):
            if clusters[i] == clusters[j]:
                data.append(1)
                row.append(i)
                col.append(j)
    return csr_matrix((data, (row, col)), shape=(n, n))

sparse_matrix_k_means = create_sparse_matrix(clusters_k_means)
sparse_matrix_mixture = create_sparse_matrix(clusters_mixture)

```

```python
sparse_matrix_mean = (sparse_matrix_k_means + sparse_matrix_mixture) / 2
sparse_matrix_mean[sparse_matrix_mean < 0.5] = 0
sparse_matrix_mean[sparse_matrix_mean >= 0.5] = 1
```

```python
#sparse_matrix_mean = sparse_matrix_mean.toarray()

clusters_final = np.zeros(len(sample))
clusters_final_next_id = 0

for i in range(len(sample)):
    if clusters_final[i] == 0:
        clusters_final_current_id = clusters_final_next_id
        clusters_final_next_id += 1
        clusters_final[i] = clusters_final_current_id
        for j in range(i+1, len(sample)):
            if sparse_matrix_mean[i, j] == 1:
                clusters_final[j] = clusters_final_current_id

```

```python
submission["Predicted"] = clusters_final
submission.to_csv("submission.csv", index=False)
submission.head()
```
